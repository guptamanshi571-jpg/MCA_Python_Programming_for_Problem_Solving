```python
n=int (input("Enter the number:"))
def bin(n):
    bin=[]
    while(n/2!=0):
        if(n%2==0):
            bin.insert(0,0)
        else:
            bin.insert(0,1)
        n=n//2
    return bin

def oct(n):
    oct=[]
    while(n/8!=0):
        r=n%8
        oct.insert(0,r)
        n=n//8
    return oct
def hexadec(n):
    hexadec=[]
    while(n/16!=0):
        r=n%16
        match r:
            case 10:
                hexadec.insert(0,'A')
            case 11:
                hexadec.insert(0,'B')
            case 12:
                hexadec.insert(0,'C')
            case 13:
                hexadec.insert(0,'D')
            case 14:
                hexadec.insert(0,'E')
            case 15:
                hexadec.insert(0,'F')
            case _:
                hexadec.insert(0,r)
        n=n//16
    return hexadec
def print_num(num):
    for n in num:
        print(n,end='')
num_oct=oct(258)
print(f"Octal representation of decimal number {n}")
print_num(num_oct)
num_bin=bin(n)
print(f"\n\nBinary representation of decimal number {n}")
print_num(num_bin)
hexa_dec=hexadec(n)
print(f"\n\nHeaxadecimal representation of decimal number {n}") 
print_num(hexa_dec)
    
```
