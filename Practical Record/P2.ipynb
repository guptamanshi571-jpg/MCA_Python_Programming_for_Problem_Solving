```python
my_string = input("Enter A String:")
print(f"Entered string is:{my_string}")
```

```python
def count_vowels(input_string):
    vowels = "aeiouAEIOU" # Define all possible vowels, both lowercase and uppercase 
    vowel_count = 0
    
    for char in input_string: 
        if char in vowels:
            vowel_count += 1
    return vowel_count
num_vowels = count_vowels (my_string)
print(f"The string '{my_string}' contains {num_vowels} vowels.")
```
