```python
# employees salary sort Program

employees = [
    {"name": 'Yukti', "salary": 45000}, 
    {"name": 'Roshni', "salary": 52000}, 
    {"name": 'Komal', "salary": 48000}
]
sorted_employees = sorted(employees, key=lambda emp: emp["salary"])

print("Sorted by Salary:")
for emp in sorted_employees:
    print(emp)
```
