```python
# Pattern 1(A)

for i in range(1, 6):  # Rows from 1 to 5
    for j in range(1, i + 1):  # Print numbers 1 to i
        print(j, end="")
    print()  # New line after each row
```

```python
rows = 4
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    print("*" *(2*i-1))
```

```python
# Pattern 1(C)

def diamond(n=3):
    
    for i in range(1, n + 1):
        print(" " * (n - i) + "*" * (2 * i - 1))
    
    for i in range(n - 1, 0, -1):
        print(" " * (n - i) + "*" * (2 * i - 1))

diamond()
```

```python
#  Pattern 1(D)

print("\n--- Pattern ---") 
rows_d = 4  
cols_d = 3   
 
for row_number in range(1, rows_d + 1): 
    for column_number in range(1, cols_d + 1): 
        if (row_number == 1 or row_number == rows_d or 
            column_number == 1 or column_number == cols_d): 
            print("* ", end="") 
        else: 
            print("  ", end="") 
    print()
```

```python
# Pattern 1(E)

n = 3

for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
